```python
import numpy as np
import matplotlib.pyplot as plt

 # Contains physical constants defined in SI units (default) as well as other systems
from scipy import constants

# Make easily readable plots with large axis labels
plt.rcParams.update({"font.size":16, "axes.labelsize":16, "font.family":"sans-serif", 
                     "font.sans-serif":"Arial"})
```

```python
e0 = 8.85e-12
u0 = 1.26e-6
q = 1.6e-19
c = 3e8

def e_field(r):
  return (1/(4*np.pi*e0)) * q/r**2

r = np.linspace(.1, 3, 100) 
E = e_field(r) 

plt.plot(r, E)
plt.xlabel('r (m)')
plt.ylabel('E (N/C)')
plt.title('Electric Field as a Function of Distance')
plt.show()
```

```python
def c_prime(f):
  return c * (1-f)

f = np.linspace(0, 1, 100)
Cp = c_prime(f)

plt.plot(f, Cp)
plt.xlabel('f ')
plt.ylabel('C prime')
plt.title('C Prime as a Function of F')
plt.show()
```

```python
def e0p(f):
  return e0/(1-f)**2

def Ep(f):
  return (1/(4*np.pi*e0p(f))) * q/(10e-6**2)

f = np.linspace(0, .99, 100)
F = e0p(f)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

ax1.plot(f, F)
ax1.set_xlabel('f')
ax1.set_ylabel('e0 prime')
ax1.set_title('e0 prime as a function of f')

EP = Ep(f)

ax2.plot(f, EP)
ax2.set_xlabel('f')
ax2.set_ylabel('E prime (N/C)')
ax2.set_title('E prime as a function of f')

plt.tight_layout()
plt.show()
```

```python
def e_prime(r_prime):
   return (1/(4*np.pi*e0)) * q/r_prime**2

def e2_prime(r_prime):
  return (1/(4*np.pi*e0p(.5))) * q/((r_prime)**2)

r_prime = np.linspace(.01, .3, 1000) 

E_p = e_prime(r_prime) # red line

plt.plot(r_prime, E_p, 'r')
plt.xlabel('r_prime (m)')
plt.ylabel('E_prime (N/C)')
plt.title('E Prime as a Function of Distance')

E2_p = e2_prime(r_prime) # blue line

plt.plot(r_prime, E2_p)
plt.xlabel('r_prime (m)')
plt.ylabel('E_prime (N/C)')
plt.title('E Prime as a Function of Distance')

plt.show()

# The problem is they are not equal.  
# Electric field cant have two values in space at one time.
# They disagree on the value of the field, 
# which can be measured to be equal.
```

```python
def mu0(f):
  return u0/(1-f)**2

f = np.linspace(.01, .99, 99)

U0 = mu0(f)

plt.plot(f, U0)
plt.xlabel('f')
plt.ylabel('mu0')
plt.title('mu0 as a Function of F')
plt.show()
```

```python
I = 1
R = 1
r = 10e-6
f_vals = np.linspace(.01, .99, 99)


def E6(f):
  return ((q+0*f)/(4*np.pi*e0*(r**2)))

def B6(f):
  return (u0*I+0*f)/(2*R)

def B_prime(f):
  return (u0*I)/(R*2*((1-f)**2))

E6vals = E6(f_vals) 
B6vals = B6(f_vals)

B_prime_vals = B_prime(f_vals)
E_prime_vals = Ep(f_vals)

plt.plot(f_vals, E6vals,label='E')
plt.plot(f_vals, E_prime_vals,label='E_prime')
plt.xlabel('f ')
plt.ylabel('E (N/C)')
plt.title('Electric Field as a Function of f')
plt.legend()
plt.show()

plt.plot(f_vals,B6vals,label='B')
plt.plot(f_vals,B_prime_vals,label='B_prime')
plt.xlabel('f')
plt.ylabel('B (A/m)')
plt.title('Magnetic Field as a Function of f')
plt.legend()
plt.show()

#  These graphs show that E PRIME = E and B PRIME = B only when f = 0,
#  which is a problem.
```
