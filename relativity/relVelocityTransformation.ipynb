```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants
```

```python
c = 3e6

def gamma(v):
  return 1/np.sqrt(1-((v**2)/(c**2)))

def vf(v):
  return (.99*c + v)/(1 + (v/c**2)*.99*c)

vm = np.linspace(1, c - 1, 100)

plt.plot(vm, vf(vm))
plt.xlabel('velocity of spacecraft (km/s)')
plt.ylabel('velocity frank sees (km/s)')
plt.title('a) Relativistic Velocity Transformation')
plt.show()

def theta_f(vm):
  gamma = 1/np.sqrt(1-(vm**2 / c**2))
  return np.arctan(gamma*np.tan(15))

def vy(vm):
  return .99*c*np.cos(15*180/np.pi)/gamma(vm)

plt.plot(vm, vy(vm))  
plt.xlabel('velocity of spacecraft (km/s)')
plt.ylabel('velocity frank sees (km/s)')
plt.title('b) Relativistic Velocity Transformation')
plt.show()
```

```python
def light(t):
  return c*t

def lighta(t):
  return -1*c*t

def light2(t):
  return c*t + c

def light2a(t):
  return -1*c*t + c

def jwt(t):
  return 1.5e6*(t/t)

def alien(t):
  return -.1*c*t + 5e6

t = np.linspace(1, 200, num=201, endpoint=True) - 100

fig, ax = plt.subplots(figsize=(7,7))


ax.set_facecolor('lightgray')
ax.plot(light(t), t*c, color="yellow", linewidth=2)
ax.plot(lighta(t), t*c, color="yellow", linewidth=2)
ax.plot(light2(t), t*c, color="yellow", linewidth=2)
ax.plot(light2a(t), t*c, color="yellow", linewidth=2)
ax.plot(jwt(t), t*c, 'b--', linewidth=2)
ax.plot(alien(t), t*c, 'r--' , linewidth=2)

ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.set_aspect('equal')

ax.set_xlim([-1e7, 1e7])
ax.set_ylim([-1e7, 1e7])

# lightbulb symbols
ax.plot(0, 0, '*', color="yellow", markersize=12)
ax.plot(3e6, 0, '*', color="yellow", markersize=12)

# jwt symbol
ax.plot(1.5e6, 0, 'b*', markersize=10)

# alien symbol
ax.plot(5e6, 0, "rx", markersize=10)

ax.set_xlabel('x (m)', labelpad=20)
ax.set_ylabel('ct (m)', rotation=0)
ax.yaxis.set_label_coords(-0.05, 0.95)
ax.xaxis.set_label_coords(1.05, .55)
ax.set_title('Spacetime')
```
