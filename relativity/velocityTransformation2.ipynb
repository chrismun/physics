```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

plt.rcParams.update({"font.size":16, "axes.labelsize":16, "font.family":"sans-serif", 
                     "font.sans-serif":"Arial"})
```

```python
def t_prime(t, v):
  return t*v

c=3e8
t_vals = np.linspace(0, 1, 1000)
v_val = .2*c

t_prime_vals = t_prime(t_vals, v_val)
plt.plot(t_vals, t_prime_vals)
plt.xlabel('t')
plt.ylabel("x")
plt.title("t' axis in the rest frame")
plt.xlim(0, 1)
plt.ylim(0, c) 
plt.show()
```

```python
def x_prime(t, v):
  return (c**2)*t/v

x_prime_vals = x_prime(t_vals, v_val)

plt.plot(t_vals, x_prime_vals)
plt.xlabel('t')
plt.ylabel("x")
plt.title("x' axis in the rest frame")
plt.xlim(0, 1) 
plt.ylim(0, c)
plt.show()
```

```python
# Set up an array of independent variables t that spans -100 to 100 seconds
t = np.linspace(0, 200, num=201, endpoint=True) - 100

# Make a figure
fig, ax = plt.subplots(figsize=(7,7))

# Plot t' axis
ax.plot(t*c, t_prime(t, 0.2*c), 'r--', linewidth=2)

# Plot x' axis
ax.plot(t*c, x_prime(t, 0.2*c), 'r--', linewidth=2)

# Move the left and bottom "spines" (borders of the plot) to the center
# This makes a math textbook-like set of (x,t) axes instead of a box around the plot
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

# Take out the top and right spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Make the two axes have the same scale
ax.set_aspect('equal')

# Set the axes ranges
ax.set_xlim([-1e10, 1e10])
ax.set_ylim([-1e10, 1e10])

ax.text(2.1e9, -1.1e9, r"$x'$", fontsize=16, color="r")
ax.text(-1.1e9, 1.1e9, r"$t'$", fontsize=16, color="r")
```

```python
# Make a figure
fig, ax = plt.subplots(figsize=(7,7))

v_list = [0.25*c, 0.5*c, 0.75*c]


color_list = ['r', 'g', 'b']

# Loop over the v values and plot the corresponding t' and x' axes
for v, color in zip(v_list, color_list):
    ax.plot(t*c, t_prime(t, v), color=color, linestyle='-', linewidth=2)
    ax.plot(t*c, x_prime(t, v), color=color, linestyle='--', linewidth=2)

# Move the left and bottom "spines" (borders of the plot) to the center
# This makes a math textbook-like set of (x,t) axes instead of a box around the plot
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

# Take out the top and right spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Make the two axes have the same scale
ax.set_aspect('equal')

# Set the axes ranges
ax.set_xlim([-1e10, 1e10])
ax.set_ylim([-1e10, 1e10])

legend_labels = [f"v={v/c:.2f}c" for v in v_list]
ax.legend(legend_labels, loc='lower right')
```

```python
# Make a figure
fig, ax = plt.subplots(figsize=(7,7))

v_list = [0.5*c, 0.75*c, 0.875*c, 0.9375*c]


color_list = [(i/5, i/5, i/5) for i in range(4, 0, -1)] # reverse order, exclusive, 1->0

# Loop over the v values and plot the corresponding t' and x' axes
for v, color in zip(v_list, color_list):
    ax.plot(t*c, t_prime(t, v), color=color, linestyle='-', linewidth=2)
    ax.plot(t*c, x_prime(t, v), color=color, linestyle='-', linewidth=2)

ax.plot(c*t, c*t, 'k-', linewidth=2, label='v = 1')

# Move the left and bottom "spines" (borders of the plot) to the center
# This makes a math textbook-like set of (x,t) axes instead of a box around the plot
ax.spines['left'].set_position(('data', 0))
ax.spines['bottom'].set_position(('data', 0))

# Take out the top and right spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# Make the two axes have the same scale
ax.set_aspect('equal')

# Set the axes ranges
ax.set_xlim([-1e10, 1e10])
ax.set_ylim([-1e10, 1e10])

v_list.append(c)

legend_labels = [f"v={v/c:.2f}c" for v in v_list]
ax.legend(legend_labels, loc='lower right')
```
