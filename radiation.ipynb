# Radiation

 Wien's displacement law, $\lambda_{\rm max} = \frac{2.898 \times 10^{-3} \mathrm{m} \cdot \mathrm{K}}{T}$, 
 
wavelength of maximum intensity for seven different types of stars: O5, B5, A5, F5, G5, K5, M5. Label both axes and each data point on your plot.  <a href="https://sites.uni.edu/morgans/astro/course/Notes/section2/spectraltemps.html">table</a> of star types and temperatures 

```python
import numpy as np
import matplotlib.pyplot as plt

k = 2.898e-3   # m K

temps = np.array([54000, 15200, 8310, 6700, 5660, 4400, 3200]) # K 

wavelengths = k / temps # m 

plt.plot(temps, wavelengths, 'ro')
plt.xlabel('Temperature (K)')
plt.ylabel('Wavelength of Maximum Intensity (m)')
plt.title("Wavelength of Maximum Intensity vs. Temperature for Seven Star Types")

star_labels = ['O5', 'B5', 'A5', 'F5', 'G5', 'K5', 'M5']
for i, label in enumerate(star_labels):
    plt.annotate(label, (temps[i], wavelengths[i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

h = 4.136e-15  # eV s
e = 1.602e-19  # C
w_Al = 4.08    # eV
w_Au = 5.10    # eV

freqs = np.linspace(0, 2.5e15, 1000)  # Hz

K_max_Al = (h * freqs - w_Al) / e   # eV
K_max_Au = (h * freqs - w_Au) / e   # eV

plt.plot(freqs, K_max_Al, label='Aluminum')
plt.plot(freqs, K_max_Au, label='Gold')
plt.axvline(x=9.88e14, linestyle='--', color='gray', label='Cutoff Frequency (Al)')
plt.axvline(x=1.23e15, linestyle='--', color='black', label='Cutoff Frequency (Au)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Maximum Kinetic Energy of Photoelectrons (eV)')
plt.title('Maximum Kinetic Energy of Photoelectrons vs. Frequency for Aluminum and Gold Photocathodes')
plt.legend()
plt.ylim(0) 
plt.show()
```
